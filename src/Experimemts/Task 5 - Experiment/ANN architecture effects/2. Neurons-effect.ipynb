### **Experiment with different ANN-Neuron size**

### **Introduction**


The number of neurons in the ANN hidden layer has a major influence on the performance of PSO trained network since it directly control the search space of the PSO must explore. Each neuron has weight and bias. A small number of neurons creates compact search space that is easier for PSO to navigate and provide more stable MAE values. When neuron count increases (16,32,64) the ANN becomes more expressive and capable of modeling more complex and may stangate in poor regions. PSO -ANN studies recommend small to moderate hidden layers(5-15)

we have done experiments with hidden layer neuron size 8,16,32

## **Results & Discussion**


This is our initial experiment used an ANN with two hidden layers which contains 16, 8,32 neurons for hidden layers

| **Hidden Layer Size** | **Activation** | **Avg gBest** | **Std gBest** | **Avg Train MAE** | **Std Train MAE** | **Avg Test MAE** | **Std Test MAE** | **Best?**                 |
| --------------------- | -------------- | ------------- | ------------- | ----------------- | ----------------- | ---------------- | ---------------- | ------------------------- |
| **8 neurons**         | ReLU           | 7.4338        | 0.3883        | 7.4338            | 0.3883            | **8.0145**       | **0.3713**       | ⭐ **Best overall**        |
| 16 neurons            | ReLU           | 7.5291        | **0.2169**    | 7.5291            | **0.2169**        | 8.0515           | **0.1640**       | Stable but not lowest MAE |
| 32 neurons            | ReLU           | **7.2734**    | 0.3306        | **7.2734**        | 0.3306            | **7.8753**       | 0.4109           | Best MAE, but less stable |


According to the experiment with different neuron size 8,16,32 hidden layers, we have observed that 32 neuron has the slightly lower MAE when compare to the 8 and 16. However, the variance is bit higher in Test MAE. since the smaller neuron size provide the competitive  MAE while maintaining the lower variance, faster convergence and greater robustness to change in hyperparameter values in PSO.

The larger architectures (16, 32, 64 neurons) increased the dimensionality of the weight space, making optimisation more difficult for PSO and introducing instability in several runs. This behaviour aligns closely with the PSO-ANN literature, which recommends small to moderate network sizes (5–15 neurons) to avoid excessive search complexity and to maintain effective swarm convergence. Therefore, based on both empirical evidence and theoretical guidance, the 8-neuron architecture was selected as the final model, as it offers the best balance of accuracy, stability, and optimisation efficiency for PSO-based training
