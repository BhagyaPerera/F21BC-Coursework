
## **4. Social Variation (gamma)**


### **Introduction**

gamma controls how strongly a particle follows the best performer in its neighbourhood (informants). Each particle has set of informants based on the method we use to select informants(eg:random_k, round topology).These informants share their best positions.Gamma tells the particle how much it should listen to these neighbours.

#### **Results & Discussion**

The findings of the three tests, which examined various PSO gamma values on an eight-neuron single-layer ANN, showed minor but significant variations in the algorithm's convergence and generalisation. With a low gamma of 0.5 and an average gBest of 7.4192 ± 0.3621, training MAE of 7.4192 ± 0.3621, and test MAE of 7.9989 ± 0.2252, the PSO showed steady but a little slower convergence. When gamma was raised to a medium value of 1.49, performance somewhat decreased, with an average gBest of 7.5546 ± 0.2934, training MAE of 7.5546 ± 0.2934, and test MAE of 8.0666 ± 0.2749. This implies that the stronger pull towards the world best may have caused some runs to converge too soon.The algorithm slightly improved training and test results (gBest 7.3925 ± 0.3276, training MAE 7.3925 ± 0.3276, and test MAE 7.8407 ± 0.3958) at a higher gamma of 2.0, but test performance showed more variability. All things considered, these results show that moderate gamma values may slightly impair performance, whereas higher gamma may enhance exploitation without significantly impacting convergence. This demonstrates the sensitivity of PSO's exploration-exploitation balance when ANN weight optimisation.


| **gamma Value**                 | **Average gBest (± SD)** | **Train MAE (± SD)** | **Test MAE (± SD)** | **Behaviour Summary**                                  | **Best?**        |
|---------------------------------| ------------------------ | -------------------- | ------------------- | ------------------------------------------------------ | ---------------- |
| **0.5** (low social influence)  | **7.4192 ± 0.3621**      | **7.4192 ± 0.3621**  | **7.9989 ± 0.2252** | More exploration, stable search, slower convergence    | **Good**         |
| **1.49** (standard PSO)         | **7.5546 ± 0.2934**      | **7.5546 ± 0.2934**  | **8.0666 ± 0.2749** | Balanced exploration + exploitation                    | Average          |
| **2.0** (high social influence) | **7.3925 ± 0.3276**      | **7.3925 ± 0.3276**  | **7.8407 ± 0.3958** | Strong convergence pressure, finds good minima quickly | **Best Overall** |


The experimental results are consistent with established PSO theory. The best performance occurred at γ = 2.0, which matches prior findings suggesting that stronger social influence (γ between 1.5–2.0) supports faster and more reliable convergence when optimising ANN weights. Lower γ values encouraged exploration but produced slightly higher MAE, also aligning with published behaviour in PSO optimisation studies.
