## **3.Experiment with different Activation functions**

### **Introduction**

An activation function is a mathemaical function applied to a neuron's output inside and artificial neural network. Its purpose is to introduce non-linearity,allowing the network to model complex patterns that cannot be captured by simple linear equations. Without activation functions ANN becomes simple linear regression model which skip complex non-linearity patterns, no matter how many layers it has.Activation functions helps ANN to learn curves,interaction between variables,non-linear relationships, threshold and saturation behaviour.

Why Activation Functions Matter in PSO + ANN?

PSO does not use gradients, so the shape of the activation function directly affects:
    the smoothness of the search landscape
    how weights influence errors
    whether neurons get “stuck” or saturate
    how stable the global best becomes
    Different activation functions produce very different optimisation behaviour.

We have used different activations such as
   -RELU
   -Tannh
   -RELU+Tanh
   -Logistics
   -Identity
