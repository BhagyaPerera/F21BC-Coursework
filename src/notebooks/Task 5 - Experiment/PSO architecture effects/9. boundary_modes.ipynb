#### **Results & Discussion**

Different performance trends were noted in the three PSO boundary mode experiments used to train the fixed ANN (32 → 16 → 8, ReLU activations). The clip mode produced the best overall results, showing good generalisation and stable convergence with a gBest of 8.058 ± 0.364, train MAE of 8.058 ± 0.364, and test MAE of 8.527 ± 0.385. The swarm's velocity was limited to [-5, 5] and its range was [-2, 2]. With a ring topology with looser bounds [-5, 5] and v_clamp [-2, 2], the invisible mode performed marginally worse (gBest 8.224 ± 0.628, test MAE 8.877 ± 0.656), indicating higher variance across runs and sporadic particle divergence.The reflect mode with very wide bounds [-10, 10] and v_clamp [-1, 1] yielded the lowest average performance and the highest variability (gBest 8.553 ± 0.617, test MAE 9.073 ± 0.603), most likely because of inconsistent exploration caused by extreme particle reflections. The most dependable and accurate optimisation was generally offered by clip mode, whereas more instability was introduced by invisible and reflect modes. This underscores the significance of boundary handling in PSO for ANN training.
