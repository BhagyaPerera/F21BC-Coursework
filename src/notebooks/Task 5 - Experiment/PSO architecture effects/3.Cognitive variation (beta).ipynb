
## **3. Cognitive Variations (beta)**


### **Introduction**

Beta controls how much a particle listens to itself.Each particle remembers the best place it has ever been (its personal best).Beta decides how strongly it tries to go back to that good spot.

Simply Beta is the particle's own confidence in its previous good experience. If beta is high, the particle strongly follows its own past success.It becomes more selfish. It explores individually.
If beta is low, the particle ignores its own experience depends on others.

### **Results & Discussion**


Overall, the cognitive coefficient beta had a moderate effect on PSO–ANN performance, with smaller beta values slightly limiting exploration and larger beta values encouraging more aggressive self-driven search. The best overall performance was observed at beta = 2.5, though the differences were small, suggesting that the PSO–ANN system is relatively robust to reasonable changes in beta.

| **beta Value**       | **Train MAE (Mean ± SD)** | **Test MAE (Mean ± SD)** | **Interpretation**                                                                                   |
|----------------------| ------------------------- | ------------------------ | ---------------------------------------------------------------------------------------------------- |
| **0.3**              | **7.5383 ± 0.2376**       | **8.0292 ± 0.2894**      | Very low cognitive pull → weak self-learning → stable but limited personal exploration               |
| **1.49** *(default)* | **7.6099 ± 0.2527**       | **8.0971 ± 0.1827**      | Balanced cognitive influence → standard behaviour → moderate error                                   |
| **2.5**              | **7.4806 ± 0.1554**       | **8.0383 ± 0.2118**      | Strong cognitive pull → more self-exploration → slightly improved convergence but higher sensitivity |


The behaviour observed in the beta-value experiments is consistent with the established findings in PSO and PSO–ANN research. Prior studies indicate that cognitive coefficients within the range of 1.0–2.5 offer a favourable balance between exploration and exploitation, enabling particles to utilise personal experience without destabilising the swarm. Our results follow this trend precisely: the low beta value (0.3) weakened self-exploration and resulted in slightly poorer MAE, whereas the higher values (beta = 1.49 and beta = 2.5) improved convergence and reduced stagnation, with beta = 2.5 producing the best performance. Thus, the experimental outcomes reinforce the theoretical expectations reported in PSO literature.
