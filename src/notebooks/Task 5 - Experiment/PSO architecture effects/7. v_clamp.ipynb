## **7. V_Clamp**

### **Introduction**

In Particle Swarm Optimization, each particle moves through the search space using a velocity vector.
v_clamp defines the allowed minimum and maximum velocity a particle can have in any dimension:
Velocity clamping (v_clamp) limits the maximum step size a particle can take during PSO updates. It prevents particle explosion, stabilises exploration, and ensures that ANN weight updates remain smooth and controlled. When v_clamp is too small, the swarm moves slowly and may stagnate; when it is too large, the swarm overshoots good regions and becomes unstable.

### **Results & Discussion**

The experiments examined the effects of PSO velocity clamping on ANN training for concrete strength prediction using a single hidden layer with eight neurons. The PSO paired with the wider velocity clamp of (-5,5) resulted with an average gBest of 7.3198 ± 0.2358, Train MAE of 7.3198 ± 0.2358, and Test MAE of 8.152 and had the best overall performance.   When the clamp was tightened to (-1,1) gBest 8.3653 ± 0.1583, Train MAE 8.3653 ± 0.1583, Test MAE 8.8014 ± 0.1493, performance further declined, and average errors marginally increased when the clamp was narrowed to (-2,2). Train MAE 7.6111 ± 0.1941, Test MAE 8.1420 ± 0.1889. This demonstrates how exploration is hampered by limiting particle velocities, which results in higher prediction errors and slower convergence.   All things considered, the findings show how crucial it is to permit adequate particle movement in PSO for successful ANN optimisation.
