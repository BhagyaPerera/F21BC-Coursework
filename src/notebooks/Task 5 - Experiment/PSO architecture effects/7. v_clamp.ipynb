#### **Results & Discussion**

The experiments examined the effects of PSO velocity clamping on ANN training for concrete strength prediction using a single hidden layer with eight neurons. The PSO paired with the wider velocity clamp of (-5,5) resulted with an average gBest of 7.758, Train MAE of 7.758, and Test MAE of 8.152 and had the best overall performance.   When the clamp was tightened to (-1,1) gBest 8.060, Train MAE 8.060, Test MAE 8.484, performance further declined, and average errors marginally increased when the clamp was narrowed to (-2,2). (Train MAE 7.766, Test MAE 8.266, gBest 7.766). This demonstrates how exploration is hampered by limiting particle velocities, which results in higher prediction errors and slower convergence.   All things considered, the findings show how crucial it is to permit adequate particle movement in PSO for successful ANN optimisation.
