#### **Results & Discussion**


These experiments train a single-layer ANN with eight neurones using PSO for various alpha values (0.1, 1.0, and 2.5), where swarm size and iterations are maintained constant at 40 and 200, respectively. With an average gBest and Train MAE of 7.437 ± 0.154 and Test MAE of 7.938 ± 0.175, the low alpha (0.1) configuration gained the best stability with slower but more consistent convergence. With slightly more variability, the average Train MAE of 7.454 ± 0.145 and the average Test MAE of 8.031 ± 0.165 for medium alpha showed a balanced trade-off between exploration and exploitation. Having a Train MAE of 7.521 ± 0.271 and a Test MAE of 8.046 ± 0.354, a high alpha (2.5) slowed down overall performance and accelerated convergence at the expense of increasing variance. In addition to showing that lower alpha values promote consistent convergence and fewer errors, these results highlight the necessity of fine-tuning cognitive coefficients in PSO while optimising ANN weights.
