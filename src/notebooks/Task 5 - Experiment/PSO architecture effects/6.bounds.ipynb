
## **6. Bounds**


### **Introduction**

In PSO, bounds define the maximum and minimum values that each particle's position can go. They effectively act as a fence, preventing particles from moving in to unrealistic regions. When bounds are too wide, particle may explore excessively large areas, making convergence slow and unstable. Bounds that are too narrow restrict the search excessively, preventing the swarm from discovering better solutions that lie outside the limited region.

| **Bounds Size**                         | **Behaviour of Particles**              | **Impact on Search**                   | **Typical Outcome**                                              |
|-----------------------------------------| --------------------------------------- | -------------------------------------- | ---------------------------------------------------------------- |
| **Very Small Bounds** (e.g., –1 to 1)   | Particle movement is heavily restricted | Very limited exploration               | Fast convergence but usually poor MAE (swarm gets stuck quickly) |
| **Medium Bounds** (e.g., –5 to 5)       | Balanced movement with enough freedom   | Good mix of exploration + exploitation | Best overall performance; stable convergence; lowest MAE         |
| **Very Large Bounds** (e.g., –10 to 10) | Particles move too widely across space  | Excessive exploration; unstable search | Slow convergence, high MAE, oscillation around solutions         |

### **Results & Discussion**

The experiment used PSO to optimize a fixed ANN with an 8-neuron ReLU hidden layer, testing weight bounds of (−2,2), (−5,5), and (−10,10) over ten runs.The most constrained bounds(-2.0, 2.0), produced the poorest result with an average Test MAE of 9.6141. While expanding the search space (-5.0, 5.0) significantly improved performance to 7.8689 , and the widest bounds(-10.0, 10.0), yielded the best average MAE of 7.5877 with less stability. (-5,5) gives the best overall performance

| **Bounds (Low, High)** | **Average Train MAE ± SD** | **Average Test MAE ± SD** | **Interpretation**                                                 |
| ---------------------- | -------------------------- | ------------------------- | ------------------------------------------------------------------ |
| **(-2, 2)**            | **8.9946 ± 0.3082**        | **9.6141 ± 0.2922**       | Bounds too tight → swarm cannot explore enough → worst performance |
| **(-5, 5)**            | **7.2363 ± 0.3132**        | **7.8689 ± 0.2812**       | **Balanced range → best overall performance**                      |
| **(-10, 10)**          | **7.1269 ± 0.4699**        | **7.5877 ± 0.5969**       | High exploration → good results but higher variance, less stable   |

Overall, the best-performing and most stable configuration for this dataset is bounds = (–5, 5), which aligns with PSO literature recommending moderate weight ranges for ANN training (Clerc & Kennedy, 2002; Van den Bergh & Engelbrecht, 2006.
